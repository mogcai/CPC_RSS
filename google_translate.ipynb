```python
from googletrans import Translator
```

```python

async def translate(sentense, from_lang, to_lang):
    translator = Translator()
    text_to_translate = await translator.translate(sentense, src=from_lang, dest=to_lang)
    return text_to_translate.text

from_lang = 'pt'
to_lang = 'en'
sentense='Os crimes de burla telefónica perpetrados sobre estudantes em 2024 totalizaram 109 casos, de um total de 354, e provocaram um prejuízo de cerca de 68 milhões de patacas. A Polícia Judiciária tem vindo a exercer um forte combate nesta matéria, tendo conseguido que os casos diminuíssem. Os casos de burla através da internet aumentaram 4% em termos anuais, para 906, sendo que o phishing foi o crime que registou o maior aumento'
```

```python
await translate(sentense, from_lang, to_lang)
```
