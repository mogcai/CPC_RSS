```python
import requests
import datetime
```

```python
def get_news(date):
    url='https://www.tdm.com.mo/api/v1.0/news'
    payload={
        'IsShowSevenDay': True,
        'Type': 'image',
        'Category': None,
        'Date': date,
        'Page': 1,
        'PageSize': 200,
        'Lang': 'zh',
        'PreRecordAd': 5,
        'Slug': 'news_list'
    }
    headers={'USER-AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36 Edg/128.0.0.0'}
    r=requests.get(url, params=payload, headers=headers)
    newList=r.json()['data']['newsList']
    return newList
```

```python
end=datetime.datetime.today()
start=datetime.datetime.strptime('2025-7-1', '%Y-%m-%d')
dates=[(end- datetime.timedelta(days=x)).strftime(format='%Y-%m-%d') for x in range((end-start).days+1)]
```

```python
news=[]
for date in dates:
    news=news+get_news(date)
```

```python
def search_news(news, lst=None):
    if lst is None:
        lst = []
    if isinstance(news, list):
        for item in news:
            search_news(item, lst)
    elif isinstance(news, dict):
        if 'data' in news:
            search_news(news['data'], lst)
        else:
            lst_id=[item['id'] for item in lst]
            if not(news['id'] in lst_id):
                lst.append(news)

    return lst
lst=search_news(news)
```

```python
news_emop=[item for item in lst if '金融' in item['content']]
[(item['title'], item['date'], item['shareUrl']) for item in news_emop]
```
