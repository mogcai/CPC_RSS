```python
import requests
```

```python
def get_news(date):
    url='https://www.tdm.com.mo/api/v1.0/news'
    payload={
        'IsShowSevenDay': True,
        'Type': 'image',
        'Category': None,
        'Date': date,
        'Page': 1,
        'PageSize': 200,
        'Lang': 'zh',
        'PreRecordAd': 5,
        'Slug': 'news_list'
    }
    headers={'USER-AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36 Edg/128.0.0.0'}
    r=requests.get(url, params=payload, headers=headers)
    newList=r.json()['data']['newsList']
    return newList
```

```python
dates=['2024-09-20', '2024-09-21', '2024-09-22', '2024-09-23', '2024-09-24', '2024-09-25']
news=[]
for date in dates:
    news=news+get_news(date)
```

```python
def search_news(news, lst=None):
    if lst is None:
        lst = []
    if isinstance(news, list):
        for item in news:
            search_news(item, lst)
    elif isinstance(news, dict):
        if 'data' in news:
            search_news(news['data'], lst)
        else:
            lst_id=[item['id'] for item in lst]
            if not(news['id'] in lst_id):
                lst.append(news)

    return lst
lst=search_news(news)
```
