```python
import requests
import pandas as pd
```

```python
def scrape_course(start, page):
    url=r'https://apps.dsedj.gov.mo/cesspublic/public/2023/get.jsp'
    # headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
    playload = {
        "type": "0",
        "org_id": "0",
        "start_date": start,
        "end_date": "",
        "key": "",
        "w0": "true",
        "w1": "true",
        "w2": "true",
        "w3": "true",
        "w4": "true",
        "w5": "true",
        "w6": "true",
        "list": "",
        "result": "",
        "content": "",
        "count": '',
        "pagination": "",
        "on_search": "false",
        "org_key": "",
        "org_show": "false",
        "org_result": "",
        "action": "search",
        "_cur_page": page
    }

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
        "Content-Type": "application/x-www-form-urlencoded"
    }

    r=requests.post(url, headers=headers, data=playload)

    return r

def search_course(start):
    r=scrape_course(start, 0)
    pages=r.json()['count']//25+2
    lst=[]
    for page in range(0,pages):
        lst+=scrape_course(start, page).json()['list']

    df=pd.DataFrame(lst)
    return df

```
